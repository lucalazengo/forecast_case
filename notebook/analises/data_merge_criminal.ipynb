```python
import matplotlib.pyplot as plt

# Remover valores nulos
dados_total = df2["total"].dropna()

# Criar a figura
plt.figure(figsize=(14,6))

# --- Histograma ---
plt.subplot(1,2,1)
plt.hist(dados_total, bins=50, edgecolor="black")
plt.title("Distribuição da coluna 'total' (todos os valores)")
plt.xlabel("Total")
plt.ylabel("Frequência")

# --- Histograma focado (para valores menores, sem outliers gigantes) ---
plt.subplot(1,2,2)
plt.hist(dados_total[dados_total <= 20], bins=20, edgecolor="black")
plt.title("Distribuição da coluna 'total' (até 20)")
plt.xlabel("Total")
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()

# --- Boxplot separado ---
plt.figure(figsize=(8,5))
plt.boxplot(dados_total, vert=False, showfliers=True)
plt.title("Boxplot da coluna 'total'")
plt.xlabel("Total")
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(10,5))
plt.hist(dados_total, bins=100, edgecolor="black", log=True)
plt.title("Histograma em escala logarítmica")
plt.xlabel("Total")
plt.ylabel("Frequência (log)")
plt.show()
```
